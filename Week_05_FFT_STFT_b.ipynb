# Week 05: FFT & STFT
## Topic 1: Use FFT to Analyze Signal Spectra

In this section, we'll explore how to use the Fast Fourier Transform (FFT) to analyze the frequency components of signals. We'll generate a sample signal with multiple frequency components and use `numpy.fft` and `scipy.fft` to compute and visualize the spectrum.

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
%matplotlib inline
```

### Step 1: Generate a Signal
We'll synthesize a signal composed of three sinusoidal waves: 10 Hz, 50 Hz, and 100 Hz.

```python
# Sampling parameters
Fs = 500  # Sampling frequency (Hz)
T = 1.0   # Duration (seconds)
t = np.linspace(0, T, int(Fs*T), endpoint=False)

# Composite signal
signal = (np.sin(2*np.pi*10*t) +
          0.5*np.sin(2*np.pi*50*t) +
          0.2*np.sin(2*np.pi*100*t))
```

### Step 2: Compute FFT and Frequency Axis

```python
# Compute FFT
fft_vals = fft(signal)
fft_mag = np.abs(fft_vals) / len(signal)  # Normalize

# Frequency axis
freqs = fftfreq(len(signal), 1/Fs)
```

### Step 3: Plot Frequency Spectrum

```python
# Plot only the positive frequencies
plt.figure(figsize=(10, 5))
plt.plot(freqs[:len(freqs)//2], fft_mag[:len(freqs)//2])
plt.title('FFT Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

# Week 05: FFT & STFT
## Topic 2: Compare DFT vs FFT Efficiency

In this section, we compare the runtime of DFT (implemented manually) and the optimized FFT algorithm. This will highlight how FFT drastically reduces computation time for large datasets.

```python
# Imports
import numpy as np
import time
from scipy.fft import fft
import matplotlib.pyplot as plt
```

### Step 1: Manual DFT Implementation

```python
def dft_manual(x):
    N = len(x)
    X = np.zeros(N, dtype=complex)
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.exp(-2j * np.pi * k * n / N)
    return X
```

### Step 2: Generate Large Test Signal

```python
N = 1024  # You can increase to 2048, 4096 for stronger contrast
t = np.linspace(0, 1, N, endpoint=False)
signal = np.sin(2*np.pi*50*t) + 0.5*np.sin(2*np.pi*120*t)
```

### Step 3: Time DFT vs FFT

```python
# Time manual DFT
start_dft = time.time()
X_dft = dft_manual(signal)
end_dft = time.time()

# Time FFT
start_fft = time.time()
X_fft = fft(signal)
end_fft = time.time()

# Display results
print(f"Manual DFT Time: {end_dft - start_dft:.4f} seconds")
print(f"FFT Time: {end_fft - start_fft:.6f} seconds")
```

### Step 4: Plotting Spectra (optional)
We plot the FFT spectrum for visualization. (Manual DFT is skipped for plotting due to inefficiency.)

```python
freq = np.fft.fftfreq(N, d=1/N)

plt.figure(figsize=(10, 5))
plt.plot(freq[:N // 2], np.abs(X_fft[:N // 2]), label='FFT Spectrum')
plt.title('FFT Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

### 🔍 Summary
- **DFT** has time complexity \( O(N^2) \), which becomes very slow for large \( N \).
- **FFT** is an optimized version of DFT with \( O(N \log N) \) complexity.
- For practical signal processing, always use `fft()` from `scipy.fft` or `numpy.fft`.

# 📘 Apply STFT for Time-Varying Signals

We use the Short-Time Fourier Transform (STFT) to analyze how frequency content evolves over time in a signal.

This is useful for non-stationary signals like speech, music, or real-world sensor data.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft
```

### Step 1: Create a Time-Varying Signal
The signal has changing frequency over time (chirp-like).

```python
fs = 1000  # Sampling frequency
T = 2      # Duration in seconds
t = np.linspace(0, T, T * fs, endpoint=False)

# Signal with changing frequency
signal = np.sin(2 * np.pi * 10 * t * (1 + t))
```

### Step 2: Compute STFT

```python
f, t_stft, Zxx = stft(signal, fs=fs, nperseg=256, noverlap=128)
```

### Step 3: Plot the Spectrogram
Magnitude of STFT (log-scaled) shows frequency over time.

```python
plt.figure(figsize=(10, 5))
plt.pcolormesh(t_stft, f, np.abs(Zxx), shading='gouraud')
plt.title('STFT Spectrogram')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.colorbar(label='Magnitude')
plt.tight_layout()
plt.show()
```

### 🔍 Summary
- STFT allows us to analyze non-stationary signals.
- It provides both time and frequency information (unlike FFT).
- Useful for speech, music, ECG, seismic data, etc.

👉 Try adjusting `nperseg` and `noverlap` to explore time-frequency resolution trade-offs.

# Apply STFT for Time-Varying Signals

Short-Time Fourier Transform (STFT) is used to analyze signals whose frequency content changes over time. It divides the signal into overlapping time segments (windows) and performs FFT on each window, giving a time-frequency representation.

## Key Concepts
- STFT provides both time and frequency resolution
- Each segment/window is multiplied with a windowing function (e.g., Hann, Hamming)
- Trade-off between time and frequency resolution: shorter windows give better time resolution, longer windows give better frequency resolution

```python
# Example: Apply STFT on a time-varying signal
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft

# Create a signal with time-varying frequency
fs = 1000  # Sampling frequency
t = np.linspace(0, 2, 2 * fs, endpoint=False)
x = np.sin(2 * np.pi * 50 * t) * (t < 1) + np.sin(2 * np.pi * 150 * t) * (t >= 1)

# Compute STFT
f, t_stft, Zxx = stft(x, fs, nperseg=256)

# Plot STFT
plt.figure(figsize=(10, 6))
plt.pcolormesh(t_stft, f, np.abs(Zxx), shading='gouraud')
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.colorbar(label='Magnitude')
plt.tight_layout()
plt.show()
```

## Discussion
- The spectrogram shows energy at 50 Hz during the first second and 150 Hz during the second.
- You can change `nperseg` to experiment with resolution trade-offs.
- Try other window types like `hann`, `hamming`, or apply zero-padding for better visual clarity.

# Create FFT-Based Spectrum Analyzer

An FFT-based spectrum analyzer transforms a signal into its frequency components, visualizing the energy distribution over frequencies. This is useful for audio processing, communications, and any domain where frequency analysis is key.

## Key Concepts
- Use FFT to convert time-domain signals into frequency domain
- Use a windowing function to reduce spectral leakage
- Compute magnitude spectrum and plot frequency vs amplitude

```python
# Spectrum analyzer using FFT
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import get_window

# Generate a signal with multiple frequencies
fs = 1000  # Sampling rate
t = np.arange(0, 1, 1/fs)
signal = np.sin(2 * np.pi * 50 * t) + 0.5 * np.sin(2 * np.pi * 120 * t)

# Apply window
window = get_window('hann', len(t))
windowed_signal = signal * window

# FFT
N = len(t)
fft_result = np.fft.fft(windowed_signal)
fft_freq = np.fft.fftfreq(N, d=1/fs)

# Only keep positive frequencies
half_N = N // 2
fft_magnitude = 2 * np.abs(fft_result[:half_N]) / N
fft_freq = fft_freq[:half_N]

# Plot the spectrum
plt.figure(figsize=(10, 5))
plt.plot(fft_freq, fft_magnitude)
plt.title('FFT-Based Spectrum Analyzer')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Discussion
- Peaks should appear near 50 Hz and 120 Hz, which are the components of the signal.
- You can enhance the analyzer with real-time plotting, smoothing filters, or a GUI.
- Try zooming in or changing the sampling rate to see resolution differences.
