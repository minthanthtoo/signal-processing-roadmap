```python
import os
#import nbformat
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, Markdown, clear_output

# Load or initialize progress
try:
    df = pd.read_csv("Learning_Tracker_progress.csv")
except FileNotFoundError:
    df = pd.DataFrame({
        "Week": list(range(1, 17)),
        "Topic": [
            "Basics: Signals & Systems", "Fourier, Z, and Laplace",
            "Sampling, Aliasing", "Digital Filters (FIR/IIR)",
            "FFT, STFT, Spectrogram", "Advanced Time-Frequency",
            "Audio & Speech", "Speech/Music Features",
            "Biomedical (ECG, EEG)", "Machine Learning + Signals",
            "Radar & Communication", "Wireless, OFDM, etc.",
            "IoT/Environmental Sensors", "Multirate + Kalman",
            "Financial Time Series", "Wavelets, ICA, Source Sep."
        ],
        "Completed": [False] * 16,
        "Hours": [0] * 16,
        "Notes": ["" for _ in range(16)]
    })

# 💡 Fix for NaN: convert Notes to strings
df["Notes"] = df["Notes"].fillna("").astype(str)

# Display title
display(Markdown("## 📘 Signal Processing Progress Tracker"))

# Lists to hold widgets
checkboxes, hour_inputs, note_inputs = [], [], []

# Create interactive widgets for each row
rows = []
header = widgets.HBox([
    widgets.HTML("<b># Topic</b>", layout=widgets.Layout(width="50%")),
    widgets.HTML("<b>✔️ Done</b>", layout=widgets.Layout(width="10%", justify_content="flex-end")),
    widgets.HTML("<b>⏱ Hours</b>", layout=widgets.Layout(width="15%", justify_content="flex-end")),
    widgets.HTML("<b>📝 Notes</b>", layout=widgets.Layout(width="25%", justify_content="flex-end")),
])
rows.append(header)

for i, row in df.iterrows():
    week_label = widgets.Label(f"{i+1}. {row['Topic']}", layout=widgets.Layout(width="50%"))
    cb = widgets.Checkbox(value=row["Completed"], layout=widgets.Layout(width="180px"))
    hr = widgets.BoundedIntText(value=row["Hours"], min=0, max=40, step=1, description="hr", layout=widgets.Layout(width="220px"), justify_content="flex-end")
    note = widgets.Text(value=row["Notes"], placeholder="Notes...", layout=widgets.Layout(width="25%"), justify_content="flex-end")
    
    checkboxes.append(cb)
    hour_inputs.append(hr)
    note_inputs.append(note)
    
    box = widgets.HBox([week_label,cb, hr, note])
    rows.append(box)
    

# Output area for summary
summary_out = widgets.Output()

# Save button
save_btn = widgets.Button(description="💾 Save Progress", button_style='success')

def on_save_clicked(b):
    total_hours = 0
    total_completed = 0
    
    for i in range(len(df)):
        df.loc[i, "Completed"] = checkboxes[i].value
        df.loc[i, "Hours"] = hour_inputs[i].value
        df.loc[i, "Notes"] = note_inputs[i].value
        total_hours += df.loc[i, "Hours"]
        total_completed += int(df.loc[i, "Completed"])
        
    df.to_csv("Learning_Tracker_progress.csv", index=False)
    
    with summary_out:
        clear_output()
        display(Markdown(f"### ✅ Weeks Completed: {total_completed} / 16 "))
        display(Markdown(f"### ⏱️ Total Hours Logged: {total_hours} hours"))

save_btn.on_click(on_save_clicked)

# Show the full table
display(widgets.VBox(rows))

display(save_btn, summary_out)
```
